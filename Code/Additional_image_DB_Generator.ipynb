```python
import glob
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import shutil
import pickle
```

```python
dest = 'Downloads/Images'
```

```python
pattern_df = pd.DataFrame(columns=np.arange(100))
pattern_df['ImName'] = None
#pattern_dict = list(np.zeros(100))
pattern_dict = {}
pattern_dict[2] = 'Negative'
pattern_names = [{} for i in range(100)]
row = 0
pattern_df
```

```python
for folder in glob.glob('../../Volumes/OTOKAM/DATA/*'):
    print(f'Opening folder: {folder[-8:]}')
    for subfolder in glob.glob(folder + '/*'):
        print(f'     Opening subfolderfolder: {subfolder[35:]}')
        screen = ''
        negativ = ''
        for file in glob.glob(subfolder + '/*'):
            if file[-10:]=='SCREEN.csv':
                screen = file
            elif file[-11:]=='NEGATIV.csv':
                negativ = file
        if negativ == '' and screen == '':
            print(f'        Folder doesnt contain useful information')
        else:
            print(f'        Extracting Images')
            if negativ!='':
                neg_df = pd.read_csv(negativ, sep=';', header=1, usecols=['ImName'])
                neg_df['ImName'] = neg_df.ImName.str.split('\\', expand=True).iloc[:, -1]
                for i in range(neg_df.shape[0]):
                    if (neg_df.loc[i, 'ImName']!='.jpg'):
                        shutil.copy(subfolder+ '/' + neg_df.loc[i, 'ImName'], dest)
                        pattern_df.loc[row, :] = 0
                        pattern_df.loc[row, 'ImName'] = neg_df.loc[i, 'ImName']
                        pattern_df.loc[row, 2] = 1
                        row += 1
                del neg_df
            if screen != '':
                screen_df = pd.read_csv(screen, sep=';', header=1, usecols=lambda x: x in ['ImName', 'Pattern', 'ScreenPattern'])
                if (screen_df.columns=='ScreenPattern').any():
                    screen_df = screen_df.rename(columns={'ScreenPattern': 'Pattern'})
                screen_df['ImName'] = screen_df.ImName.str.split('\\', expand=True).iloc[:, -1]
                screen_df = screen_df.dropna().reset_index().drop(['index'], axis=1)
                for i in range(screen_df.shape[0]):
                    if screen_df.loc[i, 'ImName']!='.jpg':
                        aux = True
                        pattern_df.loc[row, :] = 0
                        pattern_df.loc[row, 'ImName'] = screen_df.loc[i, 'ImName']
                        pattern_df.loc[row, 3] = 1
                        patterns = screen_df.loc[i, 'Pattern'].split('+')
                        for pattern in patterns:
                            if pattern[0] == '[' and pattern.find(']')!=1:
                                class_num = int(pattern[1:pattern.find(']')])
                                
                                if class_num == 44:
                                    class_num=23
                                    
                                elif class_num == 6:
                                    class_num=14
                                    
                                elif class_num == 7:
                                    class_num=15
                                    
                                elif class_num == 8:
                                    class_num = 16
                                class_name = pattern[pattern.find(']')+2:]
                                pattern_names[class_num][folder.split('/')[-1]] = class_name
                                pattern_df.loc[row, class_num] = 1
                                if class_num not in pattern_dict:
                                    pattern_dict[class_num] = class_name
                            else: 
                                row -= 1
                                aux = False
                                break
                        if aux:
                            shutil.copy(subfolder+ '/' + screen_df.loc[i, 'ImName'], dest)
                        row+=1
                del screen_df
```

```python
pattern_dict[14] = 'Homogene'
pattern_dict[15] = 'Mouchete'
```

```python
pattern_df = pattern_df.drop(pattern_df.columns[pattern_df.sum()==0], axis=1)
```

```python
pattern_df.to_csv(dest+'/image_patterns.csv')
```

```python
file_name = dest + '/pattern_dict.pkl'
open_file = open(file_name, "wb")
pickle.dump(pattern_dict, open_file)
open_file.close()
```
